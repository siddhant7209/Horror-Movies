```python
def pole(value):                                  #defining a new function for movie status
    if value >= 8.5:                              #is it worth watch movie?
        return "Blockbuster"
    elif value >= 6 and value < 8.5:
        return "Very Good"
    elif value >= 4.5 and value < 6:
        return "Average"
    elif value >= 3 and value < 4.5:
        return "Below Average"
    elif value >=0:
        return "Disaster"
    else:
        return "Ghosted"
    
```

```python
m2['Review Rating'].apply(pole)
```

```python
m2['Movie status'] = m2['Review Rating'].apply(pole)
```
